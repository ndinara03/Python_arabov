```python
# 1
x = float(input(" x: "))

if x >= 8:
    result = -x*x + x - 9
else:
    x4 = x*x*x*x
    denominator = x4 - 6

    # ,      
    if denominator == 0:
        result = ":   "
    else:
        result = 1 / denominator

#  
print("f(x) =", result)



# 2
N = int(input("   N: "))

# :     ( 10  99   -99  -10)
if ((N >= 10 and N <= 99) or (N <= -10 and N >= -99)):
    # :   2  
    if N % 2 == 0:
        print(", N    ")
    else:
        print(", N  ,  ")
else:
    print(", N    ")




# 3
x = float(input("  x: "))

#    ( math.factorial)
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

#   sin x    
def my_sin(x):
    #  x   [-2, 2]   
    pi = 3.141592653589793
    while x > 2 * pi:
        x -= 2 * pi
    while x < -2 * pi:
        x += 2 * pi

    result = 0
    #    sin x: x - x/3! + x/5! - x/7! + ...
    for n in range(10):  # 10    
        term = ((-1) ** n) * (x ** (2 * n + 1)) / factorial(2 * n + 1)
        result += term
    return result

#   cos x    
def my_cos(x):
    #  x   [-2, 2]
    pi = 3.141592653589793
    while x > 2 * pi:
        x -= 2 * pi
    while x < -2 * pi:
        x += 2 * pi

    result = 0
    #    cos x: 1 - x/2! + x/4! - x/6! + ...
    for n in range(10):
        term = ((-1) ** n) * (x ** (2 * n)) / factorial(2 * n)
        result += term
    return result

#   ln x    
def my_ln(x):
    # ln x    x > 0
    if x <= 0:
        return None

    #  x   1  : ln(1+y) = y - y/2 + y/3 - ...
    if 0.5 <= x <= 2:
        y = x - 1
        result = 0
        for n in range(1, 50):
            term = ((-1) ** (n - 1)) * (y ** n) / n
            result += term
        return result
    else:
        #   x    ln(1+y)
        # ln(x) = ln(a) + ln(x/a),  a   1
        a = 1.0
        if x > 2:
            while x > 2:
                x /= 2
                a *= 2
        else:
            while x < 0.5:
                x *= 2
                a /= 2
        return my_ln(x) + my_ln(a)

sin_val = my_sin(x)
cos_val = my_cos(x)
ln_val = my_ln(x)

#   
valid_functions = []
valid_values = []

if sin_val is not None:
    valid_functions.append("sin x")
    valid_values.append(sin_val)

if cos_val is not None:
    valid_functions.append("cos x")
    valid_values.append(cos_val)

if ln_val is not None:
    valid_functions.append("ln x")
    valid_values.append(ln_val)
else:
    print("ln x     x <= 0")

# 
n = len(valid_values)
for i in range(n):
    for j in range(0, n - i - 1):
        if valid_values[j] > valid_values[j + 1]:
            #     
            valid_values[j], valid_values[j + 1] = valid_values[j + 1], valid_values[j]
            valid_functions[j], valid_functions[j + 1] = valid_functions[j + 1], valid_functions[j]

if len(valid_values) > 0:
    print("   :")
    for i in range(len(valid_values)):
        print(f"{valid_functions[i]} = {valid_values[i]:.6f}")
else:
    print("   ")


# 4

circle_area = float(input("  : "))
square_area = float(input("  : "))

pi = 3.141592653589793
circle_radius = (circle_area / pi) ** 0.5

#   : S = a  a = S
square_side = square_area ** 0.5

# : d = a2
square_diagonal = square_side * 1.41421356237  # 2  1.41421356237

# )    ,    <=  
circle_diameter = 2 * circle_radius

if circle_diameter <= square_side:
    print(")    ")
else:
    print(")     ")

# )    ,    <=  
if square_diagonal <= circle_diameter:
    print(")    ")
else:
    print(")     ")

print(f"\n : {circle_radius:.4f}")
print(f" : {square_side:.4f}")
print(f" : {square_diagonal:.4f}")
print(f" : {circle_diameter:.4f}")




# 5
a = int(input(" a ( /, 1-8): "))
b = int(input(" b ( /, 1-8): "))
c = int(input(" c (  , 1-8): "))
d = int(input(" d (  , 1-8): "))

if not (1 <= a <= 8 and 1 <= b <= 8 and 1 <= c <= 8 and 1 <= d <= 8):
    print(":      1  8")
else:
    # )   
    #   ,        
    if a == c or b == d:
        print(")    (c, d)")
    else:
        print(")     (c, d)")

    # )   
    #   ,      
    # |a-c| = |b-d|
    diff_vertical = a - c
    diff_horizontal = b - d

    #    ()
    if diff_vertical < 0:
        abs_diff_vertical = -diff_vertical
    else:
        abs_diff_vertical = diff_vertical

    if diff_horizontal < 0:
        abs_diff_horizontal = -diff_horizontal
    else:
        abs_diff_horizontal = diff_horizontal

    if abs_diff_vertical == abs_diff_horizontal:
        print(")    (c, d)")
    else:
        print(")     (c, d)")

```

```python
# 1
print("    2x^2 + 29 (0  x  28):")

prime_numbers = []

for x in range(0, 29):
    #    
    n = 2 * x * x + 29

    #   
    is_prime = True

    if n < 2:
        is_prime = False
    elif n == 2:
        is_prime = True
    elif n % 2 == 0:
        is_prime = False
    else:
        #   
        divisor = 3
        while divisor * divisor <= n:
            if n % divisor == 0:
                is_prime = False
                break
            divisor += 2

    if is_prime:
        prime_numbers.append(n)
        print(f"{n} -  (x = {x})")
    else:
        print(f"{n} -  (x = {x})")

print(f"\n  : {len(prime_numbers)}")
print("  :", prime_numbers)



# 2
x = float(input("   x: "))

numerator = 1.0
denominator = 1.0

#    
#  : 2^n - 1,  n  1  6 (1,3,7,15,31,63)
#  : 2^n,  n  1  6 (2,4,8,16,32,64)

#   
power_of_2 = 1
for n in range(1, 7):

    power_of_2 *= 2

    # : x - (2^n - 1)
    term_numerator = x - (power_of_2 - 1)
    numerator *= term_numerator

    # : x - 2^n ( 2^0)
    term_denominator = x - power_of_2
    denominator *= term_denominator

if denominator == 0:
    print(":   ")
else:
    result = numerator / denominator
    print(f": {result:.10f}")

# 3

epsilon = 0.001

n = 1
sum_series = 0
current_factorial = 1  # 0! = 1

print("n\tn!\tn^n\ta_n\t\t")
print("-" * 50)

while True:
    # n! (  )
    if n == 1:
        current_factorial = 1
    else:
        current_factorial = current_factorial * n

    # n^n
    n_power_n = 1
    for i in range(n):
        n_power_n = n_power_n * n

    # a_n = n! / (3 * n^n)
    a_n = current_factorial / (3 * n_power_n)

    sum_series += a_n

    print(f"{n}\t{current_factorial}\t{n_power_n}\t{a_n:.8f}\t{sum_series:.8f}")

    if a_n < epsilon:
        print(f"\n : a_{n} = {a_n:.8f} < {epsilon}")
        break

    n += 1

print(f"\n : {sum_series:.8f}")



# 4

print("     (0 -  ):")

count = 0
number = int(input(" : "))
while number != 0:
    count += 1
    number = int(input(" : "))

print(f"   : {count}")
```

```python
# 1
k = int(input("k: "))

n = 1
while n <= k:
    #  
    digits = 0
    t = n
    while t > 0:
        digits += 1
        t //= 10
    s = 0
    t = n
    while t > 0:
        d = t % 10
        p = 1
        i = 0
        while i < digits:
            p *= d
            i += 1
        s += p
        t //= 10

    if s == n:
        print(n)

    n += 1





# 2

def sum_odd_facts():
    s = 0
    n = 1
    while n <= 9:
        f = 1
        i = 1
        while i <= n:
            f *= i
            i += 1
        s += f
        n += 2
    return s

print(":", sum_odd_facts())






# 3

def p_to_decimal_recursive(s, p, index=0):

    if index == len(s):
        return 0

    #   
    char = s[index]
    #    
    digit = 0
    if char == '0': digit = 0
    elif char == '1': digit = 1
    elif char == '2': digit = 2
    elif char == '3': digit = 3
    elif char == '4': digit = 4
    elif char == '5': digit = 5
    elif char == '6': digit = 6
    elif char == '7': digit = 7
    elif char == '8': digit = 8
    elif char == '9': digit = 9

    #      
    #   p    
    return digit + p * p_to_decimal_recursive(s, p, index + 1)

def p_to_decimal(s, p):
    #   
    if p < 2 or p > 9:
        print(":      2  9")
        return None

    #     
    for char in s:
        if char < '0' or char > str(p-1):
            print(f":  '{char}'   {p}- ")
            return None

    #   
    return p_to_decimal_recursive(s, p)

#  
if __name__ == "__main__":
    s = input("   p- : ")
    p = int(input("    (2-9): "))

    result = p_to_decimal(s, p)
    if result is not None:
        print(f" {s}  {p}-   {result}   ")
```
