```python
#Задание 1
x = float(input("Введите x: "))

if x >= 8:
    result = -x*x + x - 9
else:
    x4 = x*x*x*x
    denominator = x4 - 6

    # Проверяем, чтобы не было деления на ноль
    if denominator == 0:
        result = "Ошибка: деление на ноль"
    else:
        result = 1 / denominator

# Выводим результат
print("f(x) =", result)



#Задание 2
N = int(input("Введите целое число N: "))

# проверка: является ли число двузначным (от 10 до 99 или от -99 до -10)
if ((N >= 10 and N <= 99) or (N <= -10 and N >= -99)):
    # проверка: делится на 2 без остатка
    if N % 2 == 0:
        print("Да, N является четным двузначным числом")
    else:
        print("Нет, N является двузначным, но нечетным")
else:
    print("Нет, N не является двузначным числом")




#Задание 3
x = float(input("Введите число x: "))

# функция вычисления факториала (без math.factorial)
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# функция вычисления sin x с помощью ряда Тейлора
def my_sin(x):
    # привожу x к диапазону [-2π, 2π] для лучшей сходимости
    pi = 3.141592653589793
    while x > 2 * pi:
        x -= 2 * pi
    while x < -2 * pi:
        x += 2 * pi

    result = 0
    # ряд Тейлора для sin x: x - x³/3! + x⁵/5! - x⁷/7! + ...
    for n in range(10):  # 10 членов ряда для точности
        term = ((-1) ** n) * (x ** (2 * n + 1)) / factorial(2 * n + 1)
        result += term
    return result

# функция вычисления cos x с помощью ряда Тейлора
def my_cos(x):
    # привожу x к диапазону [-2π, 2π]
    pi = 3.141592653589793
    while x > 2 * pi:
        x -= 2 * pi
    while x < -2 * pi:
        x += 2 * pi

    result = 0
    # ряд Тейлора для cos x: 1 - x²/2! + x⁴/4! - x⁶/6! + ...
    for n in range(10):
        term = ((-1) ** n) * (x ** (2 * n)) / factorial(2 * n)
        result += term
    return result

# функция вычисления ln x с помощью ряда Тейлора
def my_ln(x):
    # ln x определен только для x > 0
    if x <= 0:
        return None

    # для x близких к 1 используем ряд: ln(1+y) = y - y²/2 + y³/3 - ...
    if 0.5 <= x <= 2:
        y = x - 1
        result = 0
        for n in range(1, 50):
            term = ((-1) ** (n - 1)) * (y ** n) / n
            result += term
        return result
    else:
        # для других x используем преобразование через ln(1+y)
        # ln(x) = ln(a) + ln(x/a), где a близко к 1
        a = 1.0
        if x > 2:
            while x > 2:
                x /= 2
                a *= 2
        else:
            while x < 0.5:
                x *= 2
                a /= 2
        return my_ln(x) + my_ln(a)

sin_val = my_sin(x)
cos_val = my_cos(x)
ln_val = my_ln(x)

#какие функции имеют смысл
valid_functions = []
valid_values = []

if sin_val is not None:
    valid_functions.append("sin x")
    valid_values.append(sin_val)

if cos_val is not None:
    valid_functions.append("cos x")
    valid_values.append(cos_val)

if ln_val is not None:
    valid_functions.append("ln x")
    valid_values.append(ln_val)
else:
    print("ln x не имеет смысла при x <= 0")

#пузырьковая сортировка
n = len(valid_values)
for i in range(n):
    for j in range(0, n - i - 1):
        if valid_values[j] > valid_values[j + 1]:
            #меняю местами значения и названия функций
            valid_values[j], valid_values[j + 1] = valid_values[j + 1], valid_values[j]
            valid_functions[j], valid_functions[j + 1] = valid_functions[j + 1], valid_functions[j]

if len(valid_values) > 0:
    print("Значения в порядке возрастания:")
    for i in range(len(valid_values)):
        print(f"{valid_functions[i]} = {valid_values[i]:.6f}")
else:
    print("Нет значений для сравнения")


#Задание 4

circle_area = float(input("Введите площадь круга: "))
square_area = float(input("Введите площадь квадрата: "))

pi = 3.141592653589793
circle_radius = (circle_area / pi) ** 0.5

# вычисляю сторону квадрата: S = a² ⇒ a = √S
square_side = square_area ** 0.5

#диагональ квадрата: d = a√2
square_diagonal = square_side * 1.41421356237  # √2 ≈ 1.41421356237

# а) круг уместится в квадрате, если диаметр круга <= стороне квадрата
circle_diameter = 2 * circle_radius

if circle_diameter <= square_side:
    print("а) Круг уместится в квадрате")
else:
    print("а) Круг НЕ уместится в квадрате")

# б) квадрат уместится в круге, если диагональ квадрата <= диаметру круга
if square_diagonal <= circle_diameter:
    print("б) Квадрат уместится в круге")
else:
    print("б) Квадрат НЕ уместится в круге")

print(f"\nРадиус круга: {circle_radius:.4f}")
print(f"Сторона квадрата: {square_side:.4f}")
print(f"Диагональ квадрата: {square_diagonal:.4f}")
print(f"Диаметр круга: {circle_diameter:.4f}")




#Задание 5
a = int(input("Введите a (вертикаль ладьи/слона, 1-8): "))
b = int(input("Введите b (горизонталь ладьи/слона, 1-8): "))
c = int(input("Введите c (вертикаль целевого поля, 1-8): "))
d = int(input("Введите d (горизонталь целевого поля, 1-8): "))

if not (1 <= a <= 8 and 1 <= b <= 8 and 1 <= c <= 8 and 1 <= d <= 8):
    print("Ошибка: все числа должны быть от 1 до 8")
else:
    # а) Проверка для ладьи
    # ладья угрожает полю, если оно на той же вертикали или горизонтали
    if a == c or b == d:
        print("а) Ладья угрожает полю (c, d)")
    else:
        print("а) Ладья НЕ угрожает полю (c, d)")

    # б) проверка для слона
    # слон угрожает полю, если оно на той же диагонали
    # |a-c| = |b-d|
    diff_vertical = a - c
    diff_horizontal = b - d

    # использую абсолютное значение (модуль)
    if diff_vertical < 0:
        abs_diff_vertical = -diff_vertical
    else:
        abs_diff_vertical = diff_vertical

    if diff_horizontal < 0:
        abs_diff_horizontal = -diff_horizontal
    else:
        abs_diff_horizontal = diff_horizontal

    if abs_diff_vertical == abs_diff_horizontal:
        print("б) Слон угрожает полю (c, d)")
    else:
        print("б) Слон НЕ угрожает полю (c, d)")

```

```python
#Задание 1
print("Простые числа по формуле 2x^2 + 29 (0 ≤ x ≤ 28):")

prime_numbers = []

for x in range(0, 29):
    # Вычисляем число по формуле
    n = 2 * x * x + 29

    # Проверяем на простоту
    is_prime = True

    if n < 2:
        is_prime = False
    elif n == 2:
        is_prime = True
    elif n % 2 == 0:
        is_prime = False
    else:
        # Проверяем нечетные делители
        divisor = 3
        while divisor * divisor <= n:
            if n % divisor == 0:
                is_prime = False
                break
            divisor += 2

    if is_prime:
        prime_numbers.append(n)
        print(f"{n} - простое (x = {x})")
    else:
        print(f"{n} - составное (x = {x})")

print(f"\nНайдено простых чисел: {len(prime_numbers)}")
print("Список простых чисел:", prime_numbers)



#Задание 2
x = float(input("Введите действительное число x: "))

numerator = 1.0
denominator = 1.0

# вычисляю через степени двойки
# для числителя: 2^n - 1, где n от 1 до 6 (1,3,7,15,31,63)
# для знаменателя: 2^n, где n от 1 до 6 (2,4,8,16,32,64)

# вычисляю степени двойки
power_of_2 = 1
for n in range(1, 7):

    power_of_2 *= 2

    # числитель: x - (2^n - 1)
    term_numerator = x - (power_of_2 - 1)
    numerator *= term_numerator

    # знаменатель: x - 2^n (кроме 2^0)
    term_denominator = x - power_of_2
    denominator *= term_denominator

if denominator == 0:
    print("Ошибка: деление на ноль")
else:
    result = numerator / denominator
    print(f"Результат: {result:.10f}")

#Задание 3

epsilon = 0.001

n = 1
sum_series = 0
current_factorial = 1  # 0! = 1

print("n\tn!\tn^n\ta_n\t\tСумма")
print("-" * 50)

while True:
    # n! (использую предыдущее значение)
    if n == 1:
        current_factorial = 1
    else:
        current_factorial = current_factorial * n

    # n^n
    n_power_n = 1
    for i in range(n):
        n_power_n = n_power_n * n

    # a_n = n! / (3 * n^n)
    a_n = current_factorial / (3 * n_power_n)

    sum_series += a_n

    print(f"{n}\t{current_factorial}\t{n_power_n}\t{a_n:.8f}\t{sum_series:.8f}")

    if a_n < epsilon:
        print(f"\nУсловие достигнуто: a_{n} = {a_n:.8f} < {epsilon}")
        break

    n += 1

print(f"\nИтоговая сумма: {sum_series:.8f}")



#Задание 4

print("Введите набор ненулевых целых чисел (0 - признак конца):")

count = 0
number = int(input("Введите число: "))
while number != 0:
    count += 1
    number = int(input("Введите число: "))

print(f"Количество чисел в наборе: {count}")
```

```python
#Задание 1
k = int(input("k: "))

n = 1
while n <= k:
    # Количество цифр
    digits = 0
    t = n
    while t > 0:
        digits += 1
        t //= 10
    s = 0
    t = n
    while t > 0:
        d = t % 10
        p = 1
        i = 0
        while i < digits:
            p *= d
            i += 1
        s += p
        t //= 10

    if s == n:
        print(n)

    n += 1





#Задание 2

def sum_odd_facts():
    s = 0
    n = 1
    while n <= 9:
        f = 1
        i = 1
        while i <= n:
            f *= i
            i += 1
        s += f
        n += 2
    return s

print("Результат:", sum_odd_facts())






#Задание 3

def p_to_decimal_recursive(s, p, index=0):

    if index == len(s):
        return 0

    # получаем текущую цифру
    char = s[index]
    # преобразуем символ в число
    digit = 0
    if char == '0': digit = 0
    elif char == '1': digit = 1
    elif char == '2': digit = 2
    elif char == '3': digit = 3
    elif char == '4': digit = 4
    elif char == '5': digit = 5
    elif char == '6': digit = 6
    elif char == '7': digit = 7
    elif char == '8': digit = 8
    elif char == '9': digit = 9

    # рекурсивный вызов для оставшейся части строки
    # умножаем на p и добавляем текущую цифру
    return digit + p * p_to_decimal_recursive(s, p, index + 1)

def p_to_decimal(s, p):
    # проверка корректности основания
    if p < 2 or p > 9:
        print("Ошибка: основание системы должно быть от 2 до 9")
        return None

    # проверка корректности цифр в строке
    for char in s:
        if char < '0' or char > str(p-1):
            print(f"Ошибка: цифра '{char}' недопустима в {p}-ичной системе")
            return None

    # вызов рекурсивной функции
    return p_to_decimal_recursive(s, p)

# основная программа
if __name__ == "__main__":
    s = input("Введите число в p-ичной системе: ")
    p = int(input("Введите основание системы счисления (2-9): "))

    result = p_to_decimal(s, p)
    if result is not None:
        print(f"Число {s} в {p}-ичной системе равно {result} в десятичной системе")
```
